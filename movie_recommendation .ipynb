```python
import numpy as np 
import pandas as pd 
import warnings
```

```python
warnings.filterwarnings('ignore');
```

```python
colomn_names = ['user id' , 'item id' , 'rating' , 'timestamp']
table1  = pd.read_csv('u.data', sep='\t', names = colomn_names);
print(table1.head())
```

```python
#df['user id'].nunique()
table2  = pd.read_csv('u.item' , sep ='\|', header = None); 
table2.shape
```

```python
# we just need the movies name 
movies_titles = table2[[0,1]]
movies_titles.columns = ['item id' , 'titles']
movies_titles
```

```python
#merging table1 and movies titles 
df = pd.merge(table1,movies_titles,on="item id")
df
```

# exploratory data analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.groupby('titles').mean()['rating'].sort_values(ascending = False).head()
```

```python
df.groupby('titles').count()['rating'].sort_values(ascending = False)
```

```python
# data frame of titles and ratings ..
ratingsDF = pd.DataFrame(df.groupby('titles').mean()['rating'])

#adding number of rating colomn in dataframe 
ratingsDF['number of rating'] = df.groupby('titles').count()['rating']
ratingsDF
```

```python
df.head()
```

```python
# create a pivot matrix 
moviematrix = df.pivot_table(index = 'user id',columns='titles',values='rating')
moviematrix
```

```python
# pass movie 
watched_movie = moviematrix['Star Wars (1977)'];

similar_to_watched_movie = moviematrix.corrwith(watched_movie)
corelation_with_watched_Movie = pd.DataFrame(similar_to_watched_movie , columns=["correlation"])
```

```python
#remove NAN values from dataframes 
corelation_with_watched_Movie.dropna(inplace=True)
corelation_with_watched_Movie
```

```python
corelation_with_watched_Movie.sort_values('correlation', ascending=False)
```

''' in above data frame the correlation is based on the only rating but
Hollow Reed (1996) this movie has not rated by lot people hence 
we need to consider 'number of count people has rated '''

```python
corelation_with_watched_Movie = corelation_with_watched_Movie.join(ratingsDF)
corelation_with_watched_Movie
```

```python
corelation_with_watched_Movie[corelation_with_watched_Movie['number of rating'] > 100].sort_values('correlation' ,ascending = False)
```
